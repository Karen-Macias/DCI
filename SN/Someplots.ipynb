```python
%pylab inline 
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.uniform(-1,1,5000)
y=np.random.uniform(-1,1,5000)
z=np.random.uniform(-1,1,5000)
```

```python
th=np.random.uniform(0.,2*np.pi,5000)
r1=np.sqrt(np.random.uniform(0,1,5000))
r2=np.random.uniform(0,1,5000)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.scatter(r1*np.cos(th),r1*np.sin(th),color='b',s=0.75)
ax.scatter(x,y,color='r',s=0.75)
ax.set_aspect('equal')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)


```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.scatter(r2*np.cos(th),r2*np.sin(th),color='b',s=0.75)
ax.scatter(x,y,color='r',s=0.75)
ax.set_aspect('equal')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)



```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
n = 100
ax.scatter(x, y, z)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```

```python
th=np.random.uniform(0.,np.pi,1000)
phi=np.random.uniform(0.,2*np.pi,1000)
rr=np.random.uniform(0,1,1000)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
n = 100
ax.scatter(rr*np.sin(th)*np.cos(phi), rr*np.sin(th)*np.sin(phi), rr*np.cos(th))
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```

```python
th=np.random.uniform(0.,2*np.pi,1000)
phi=np.arccos(np.random.uniform(-1,1,1000))
rr=np.sqrt(np.random.uniform(0,1,1000))
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
n = 100
ax.scatter(rr*np.cos(th)*np.sin(phi), rr*np.sin(th)*np.sin(phi), rr*np.cos(phi))
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
```
